```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from urllib import parse
import re
import requests
```

```python
URL = "https://namu.wiki/w/%EB%8F%99%EB%AC%BC%EC%9D%98%20%EC%88%B2%20%EC%8B%9C%EB%A6%AC%EC%A6%88/%EA%B3%A4%EC%B6%A9"
def get_html(url):
    r = requests.get(url)
    return r.text
html = get_html(URL)
soup=BeautifulSoup(html,"html.parser")
```

```python
t = soup.find_all("table")
print(len(t))
```

```python
col = t[3].tr # 실제 곤충 데이터가 있는 테이블 설정
cols = col.select('strong')
print(cols)
```

```python
columns = []
for column in cols:
    columns.append(column.text)
print(columns[:-1])
```

```python
df = pd.DataFrame(columns = columns[:-1])
print(df) # 칼럼만 있는 데이터 생성
```

```python
content = t[3] # 민물고기 목록
insects = content.select('td')
# print(content)
insectlist = []
for insect in insects:
    insectlist.append(insect.text)
print(insectlist)
```

```python
insectnum = len(insectlist)//7 - 1
print(insectnum)
for j in range(2, insectnum+2): # 0-34
    print(insectlist[(j-1)*7:((j-1)*7)+6 ])
    df.loc[j-2] = insectlist[(j-1)*7:((j-1)*7)+6 ] # 0 6 7 13
print(df.tail())
```
